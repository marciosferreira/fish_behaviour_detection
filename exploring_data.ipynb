```python
concatenated_df["tail_uniformity"] = np.NaN

def vari(the_tuple):
    if isinstance(the_tuple, tuple):
        y_tuple = tuple(zip(*the_tuple))[1]
        cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100        
        result = cv(y_tuple)
        if result is not None:      
            return result
        else:
            return np.NaN
    else:
        np.NaN

concatenated_df["tail_uniformity"] = concatenated_df.angle_corr_tail.apply(vari)
```

```python
concatenated_df["tail_crescent"] = np.NaN

def tail_cresc(the_tuple):
    if isinstance(the_tuple, tuple):
        x_tuple = tuple(zip(*the_tuple))[0]
        if sorted(x_tuple) == x_tuple:             
            return 1
        else:
            return 0
    else:
        np.NaN

concatenated_df["tail_crescent"] = concatenated_df.angle_corr_tail.apply(tail_cresc)
```
